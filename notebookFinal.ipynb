```python
def Sampling(data, k, centers, Sample_size):
    # Calculate the distance from each point in data to the nearest center
    distances = np.min(cdist(data, centers, 'euclidean')**2, axis=1)
    phi = np.sum(distances) / len(data)
    
    # Calculate the sampling probabilities for each point
    probabilities = distances / phi
    
    # Normalize the probabilities
    probabilities /= np.sum(probabilities)
    
    # Sample points based on the calculated probabilities
    sample_indices = np.random.choice(len(data), size=Sample_size, p=probabilities)
    coreset = data[sample_indices]
    
    # Calculate the weights for the sampled points
    weight = 1 / (Sample_size * probabilities[sample_indices])
    
    return coreset, weight
```
