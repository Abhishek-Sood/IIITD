```python
import numpy as np

# # Modified
# def Sampling(data, k, centers, Sample_size):
#     # Calculate distances between data points and centers
#     distances = np.min(np.linalg.norm(data[:, np.newaxis] - centers, axis=2)**2, axis=1)
    
#     # Calculate sampling probabilities
#     phi = np.sum(distances) / len(data)
#     probabilities = distances / phi
    
#     # Normalize probabilities
#     probabilities /= np.sum(probabilities)
    
#     # Sample points and calculate weights
#     coreset_indices = np.random.choice(len(data), size=Sample_size, replace=False, p=probabilities)
#     coreset_weights = 1 / (Sample_size * probabilities[coreset_indices])
#     coreset_data_points = data[coreset_indices]
    
#     return coreset_data_points, coreset_weights

def Sampling(data, k, centers, Sample_size):
    # Initialize arrays to store distances and weights
    distances = np.zeros(len(data))
    weights = np.zeros(len(data))
    
    # First loop: Calculate the distance from each point in data to the nearest center
    for i in range(len(data)):
        min_distance = float('inf')
        for j in range(len(centers)):
            distance = np.linalg.norm(data[i] - centers[j])**2
            if distance < min_distance:
                min_distance = distance
        distances[i] = min_distance
    
    phi = np.sum(distances) / len(data)
    
    # Second loop: Calculate the sampling probabilities for each point
    probabilities = np.zeros(len(data))
    
    for i in range(len(distances)):
        probabilities[i] = distances[i] / phi
    
    # Normalize the probabilities
    probabilities /= np.sum(probabilities)
    
    # Third loop: Sample points based on calculated probabilities and calculate their weights 
    coreset_indices = []
    coreset_weights = []
    
    for _ in range(Sample_size):
        index = np.random.choice(len(data), p=probabilities)
        coreset_indices.append(index)
        weight = 1 / (Sample_size * probabilities[index])
        coreset_weights.append(weight)

    coreset_data_points = data[coreset_indices]
    
    return coreset_data_points, coreset_weights
```

```python
coreset_data_points, coreset_weights = Sampling(data, k, centers, Sample_size)
print("Coreset shape:", coreset_data_points.shape)
print("Weights:", coreset_weights)
```

```python
# Calculate the clustering cost of the original KMeans model
fkmeans = KMeans(n_clusters=k, init='k-means++')
fkmeans.fit(data)
optimal_cost = fkmeans.inertia_

# Calculate Practical Coreset Relative Error
Coreset_centers, _ = wkpp.kmeans_plusplus_w(coreset_data_points, k, w=coreset_weights, n_local_trials=100)
wt_kmeansclus = KMeans(n_clusters=k, init=Coreset_centers, max_iter=10).fit(coreset_data_points, sample_weight=coreset_weights)
coreset_cost = np.sum(np.min(cdist(data, wt_kmeansclus.cluster_centers_)**2, axis=1))
relative_error_practicalCoreset = abs(coreset_cost - optimal_cost) / optimal_cost

# Calculate Uniform Sampling Coreset Relative Error
uniform_centers = KMeans(n_clusters=k, init='k-means++', max_iter=10).fit(coreset_data_points, sample_weight=coreset_weights).cluster_centers_
uniform_cost = np.sum(np.min(cdist(data, uniform_centers)**2, axis=1))
relative_error_uniformCoreset = abs(uniform_cost - optimal_cost) / optimal_cost

print("Relative error from Practical Coreset is", relative_error_practicalCoreset)
print("Relative error from Uniformly random Coreset is", relative_error_uniformCoreset)
```
